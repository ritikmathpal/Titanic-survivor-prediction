# Defining a column transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer(transformers=[("num",StandardScaler(),["Age"]),("ohe",OneHotEncoder(),["Family type","Pclass"])
                                             ,("oe",OrdinalEncoder(),["Sex"])],remainder="drop")
```

# Defining the model


```python
class LogisticRegression():

  #declaring learning rate and number of iterations(hyperparameters)
  def __init__(self,learning_rate,no_of_iterations):
    self.learning_rate=learning_rate
    self.no_of_iterations=no_of_iterations

  #fit the model to training dataset
  def fit(self,X,Y):

    #number of rows and columns in the dataset except dependent variable
    self.m, self.n=X.shape

    #initial guess for weights and bias
    self.w=np.zeros(self.n)
    self.b=0

    self.X=X
    self.Y=Y

    #implementing gradient descent
    for i in range(self.no_of_iterations):
      self.updating_weights()


  def updating_weights(self):
    #calculating y_hat the sigmoid funtion
    y_hat=1/(1+np.exp(-(self.X.dot(self.w)+self.b )))

    #derivatives
    dw=(1/self.m)*(np.dot(self.X.T,y_hat-self.Y))
    db=(1/self.m)*np.sum(y_hat-self.Y)

    #updating the weigths
    self.w=self.w-self.learning_rate*dw
    self.b=self.b-self.learning_rate*db

  def prediction(self,X):
    y_hat=1/(1+np.exp(-(X.dot(self.w)+self.b )))
    y_hat=np.where(y_hat>0.5,1,0)
    return y_hat

  def accuracy(self,prediction,Y):
      correct=0
      for i in range(Y.shape[0]):
          if prediction[i]==Y[i]:
              correct+=1
      return correct/Y.shape[0]

  def precision(self,prediction,Y):
      tp=0
      fp=0
      for i in range(Y.shape[0]):
          if prediction[i]==1:
              if Y[i]==1:
                  tp+=1
              elif Y[i]==0:
                  fp+=1
      return tp/(tp+fp)
```

# Fitting the data to the model

```python
classifier=LogisticRegression(0.1,1000)
```

```python
classifier.fit(x_train_final,y_train)
```

```python
classifier.w,classifier.b
```

# Model acurracy on training set

```python
x_train_prediction=classifier.prediction(x_train_final)
```

```python
training_accuracy=classifier.accuracy(x_train_prediction,y_train.to_numpy())
print(f"The model accuracy on training data is {training_accuracy *100}%")
```

# Model accuracy on test set

```python
x_test_final=preprocessor.fit_transform(x_test)
```

```python
x_test_prediction=classifier.prediction(x_test_final)
```

```python
test_accuracy=classifier.accuracy(x_test_prediction,y_test.to_numpy())
print(f"The model accuracy on training data is {test_accuracy *100}%")
```

# Model precision

```python
training_precision=classifier.precision(x_train_prediction,y_train.to_numpy())
print(f"The model precsion on test set is {training_precision*100}%")
```

```python
test_precision=classifier.precision(x_test_prediction,y_test.to_numpy())
print(f"The model precsion on test set is {test_precision*100}%")
```
